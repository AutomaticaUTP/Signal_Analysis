# Análisis de forma de onda periódica

## simetría de la forma de onda.

Una señal se define como par si cumple la siguiente relación.

$$
x(t) = x(-t)
$$

```
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from scipy.signal import sawtooth

tn = np.linspace(-3, 3, 1000)

figure, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 8))

for row in axes:
    for col in row:
        col.xaxis.set_label_coords(1, .35)
        col.set_xlabel('t')
        col.spines['left'].set_position('zero')
        col.spines['bottom'].set_position('zero')
        col.spines['right'].set_color('none')
        col.spines['top'].set_color('none')
        col.set_xticks([])
        col.set_yticks([])
        col.set_ylim(-1, 1.5)


styles1 = "->, head_length=5, head_width=5"
styles2 = "->, head_length=1, head_width=1"

axes[0, 0].plot(tn, -0.5*sawtooth(2 * np.pi * 0.5 * tn, 0.5) + 1)
axes[0, 0].set_title("$x(t)$")
axes[0, 0].add_patch(
    patches.FancyArrowPatch((-1, -0.5), (1, -0.5), arrowstyle=styles1, connectionstyle="arc3,rad=.5"))

axes[0, 1].plot(tn, -0.5*sawtooth(2 * np.pi * 0.5 * -tn, 0.5) + 1)
axes[0, 1].set_title("$x(-t)$")
axes[0, 1].text(
    1, -0.5, "Señal Par", size=20, ha="center", va="center", bbox=dict(boxstyle="round", fc='white',))

figure.add_artist(
    patches.ConnectionPatch(xyA=(1.5, -.5 ), coordsA=axes[0,0].transData,
                      xyB=(-1.5, -.5), coordsB=axes[0, 1].transData, arrowstyle=styles2))

axes[1, 0].plot(tn, -0.5*sawtooth(2 * np.pi * tn, 0) + 1)
axes[1, 0].set_title("$x(t)$")
axes[1, 0].add_patch(
    patches.FancyArrowPatch((-1, -0.5), (1, -0.5), arrowstyle=styles1, connectionstyle="arc3,rad=.5"))


axes[1, 1].plot(tn, -0.5*sawtooth(2 * np.pi * -tn, 0) + 1)
axes[1, 1].set_title("$x(-t)$")
axes[1, 1].text(
    1, -0.5, "Señal No Par", size=20, ha="center", va="center", bbox=dict(boxstyle="round", fc='white',))

figure.add_artist(
    patches.ConnectionPatch(xyA=(1.5, -.5 ), coordsA=axes[1,0].transData,
                      xyB=(-1.5, -.5), coordsB=axes[1, 1].transData, arrowstyle=styles2))


figure.show()
```

Como se observa una señal par es simétrica con respecto al eje vertical.

Una señal se define como impar, cuando cumple la siguiente relación.

$$
x(t) = -x(-t)
$$

```
figure = plt.figure(figsize=(15, 8))
ax1 = figure.add_axes([0, 0.5, 0.45, 0.45])
ax2 = figure.add_axes([0.5, 0.5, 0.45, 0.45])
ax3 = figure.add_axes([0.5, 0, 0.45, 0.45])

for ax in figure.get_axes():
    ax.xaxis.set_label_coords(1, .35)
    ax.set_xlabel('t')
    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_position('zero')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_ylim(-1.5, 1.5)

ax1.plot(tn, sawtooth(np.pi * tn + np.pi, 1))
ax1.set_title("$x(t)$")
ax1.add_patch(
    patches.FancyArrowPatch((-.5, -1), (.5, -1), arrowstyle=styles1, connectionstyle="arc3,rad=.5"))

figure.add_artist(
    patches.ConnectionPatch(xyA=(2.5, -1.2 ), coordsA=ax1.transData,
                      xyB=(-2.5, -1.2), coordsB=ax2.transData, arrowstyle=styles2))

ax2.plot(tn, sawtooth(-np.pi * tn + np.pi, 1))
ax2.set_title("$x(-t)$")
ax2.add_patch(
    patches.FancyArrowPatch((3, -.5), (3, .5), arrowstyle=styles1, connectionstyle="arc3,rad=.5"))

figure.add_artist(
    patches.ConnectionPatch(xyA=(2.5, -1.2 ), coordsA=ax2.transData,
                      xyB=(2.425, 1.2), coordsB=ax3.transData, arrowstyle=styles2))

ax3.plot(tn, -sawtooth(-np.pi * tn + np.pi, 1))
ax3.set_title("$-x(-t)$")
```

Por lo tanto la función $x(t)$ es una función impar.

Como se observa una señal impar es antisimétrica respeto al eje vertical en el origen.

De lo anterior puede demostrarse que:



*   $f_{par}(t) \cdot f_{par}(t) = f_{par}(t)$

*   $f_{par}(t) \cdot f_{impar}(t) = f_{impar}(t)$

*   $f_{impar}(t) \cdot f_{par}(t) = f_{impar}(t)$


*   $f_{imppar}(t) \cdot f_{impar}(t) = f_{impar}(t)$


Demostrar que si una señal es par:

$$
\int_{-a}^{a} x_{par}(t)~dt = 2\int_{0}^{a} x_{par}(t)~dt
$$

Solución:

$$
\int_{-a}^{a} x_{par}(t)~dt = \int_{-a}^{0} x_{par}(t)~dt + \int_{0}^{a} x_{par}(t)~dt
$$

Sea $t = -\tau$, $dt = -d\tau$. Recordande que $x(t) = x(-t)$.

$$
\int_{-a}^{a} x_{par}(t)~dt = -\int_{a}^{0} x_{par}(-\tau)~d\tau + \int_{0}^{a} x_{par}(t)~dt
$$

$$
\int_{-a}^{a} x_{par}(t)~dt = \int_{0}^{a} x_{par}(-\tau)~d\tau + \int_{0}^{a} x_{par}(t)~dt
$$

$$
\int_{-a}^{a} x_{par}(t)~dt = \int_{0}^{a} x_{par}(\tau)~d\tau + \int_{0}^{a} x_{par}(t)~dt
$$

$$
\int_{-a}^{a} x_{par}(t)~dt = 2\int_{0}^{a} x_{par}(t)~dt
$$

Demostrar que si una señal es impar:

$$
\int_{-a}^{a} x_{impar}(t)~dt = 0
$$

Solución:

$$
\int_{-a}^{a} x_{impar}(t)~dt = \int_{-a}^{0} x_{impar}(t)~dt + \int_{0}^{a} x_{impar}(t)~dt
$$

Sea $t = -\tau$, $dt = -d\tau$. Recordande que $x(t) = -x(-t)$.

$$
\int_{-a}^{a} x_{impar}(t)~dt = -\int_{a}^{0} -x_{impar}(-\tau)~d\tau + \int_{0}^{a} x_{impar}(t)~dt
$$

$$
\int_{-a}^{a} x_{impar}(t)~dt = \int_{a}^{0} x_{impar}(-\tau)~d\tau + \int_{0}^{a} x_{impar}(t)~dt
$$

$$
\int_{-a}^{a} x_{impar}(t)~dt = -\int_{0}^{a} x_{impar}(\tau)~d\tau + \int_{0}^{a} x_{impar}(t)~dt
$$

$$
\int_{-a}^{a} x_{impar}(t)~dt = 0
$$

Podemos además demostrar que una señal cualquiera que no sea par ni impar, puede representarse como la suma de una función par y otra impar.

$$
x(t) = x_{par}(t) + x_{impar}(t) \tag{1}
$$

$$
x(-t) = x_{par}(t) - x_{impar}(t) \tag{2}
$$

Sumando $(1)$ y $(2)$.

$$
2 \cdot x_{par}(t) = x(t) + x(-t) \\
x_{par}(t) = \frac{x(t) + x(-t)}{2}
$$

Restando a $(1)$ $(2)$.

$$
2 \cdot x_{impar}(t) = x(t) - x(-t) \\
x_{impar}(t) = \frac{x(t) - x(-t)}{2}
$$

Con relación a la simetría de una señal par ¿Cómo serán sus armónicos utilizando el espacio trigonométrio de Fourier?

Si $x(t)$ es par:

$$
a_0 = \frac{1}{T} \int_{-T/2}^{T/2} x_{par}(t)~dt = \frac{2}{T} \int_{0}^{T/2} x_{par}(t)~dt
$$

$$
a_n = \frac{2}{T} \int_{-T/2}^{T/2} x_{par}(t)\cdot cos(\omega_0 \cdot n \cdot t)~dt = \frac{4}{T} \int_{0}^{T/2} x_{par}(t)\cdot cos(\omega_0 \cdot n \cdot t)~dt
$$

$$
b_n = \frac{2}{T} \int_{-T/2}^{T/2} x_{par}(t)\cdot sen(\omega_0 \cdot n \cdot t)~dt = 0
$$

Si $x(t)$ es impar:

$$
a_0 = \frac{1}{T} \int_{-T/2}^{T/2} x_{impar}(t)~dt = 0
$$

$$
a_n = \frac{2}{T} \int_{-T/2}^{T/2} x_{impar}(t)\cdot cos(\omega_0 \cdot n \cdot t)~dt = 0
$$

$$
b_n = \frac{2}{T} \int_{-T/2}^{T/2} x_{impar}(t)\cdot sen(\omega_0 \cdot n \cdot t)~dt = \frac{4}{T} \int_{0}^{T/2} x_{par}(t)\cdot sen(\omega_0 \cdot n \cdot t)~dt
$$

## Señales con simetría de media onda.

Se dice que una función $x(t)$, cuenta con simetría de media onda cuando la siguiente relación se cumple.

$$
x(t) = -x\left(t \pm \frac{T}{2}\right)
$$

Donde $T$ es el periodo de la señal $x(t)$.

La señal $x(t)$ que se muestra, cuenta con simetría de media onda. Nótese que ella se repite cada medio ciclo pero reflejado.


```
tn = np.linspace(-1, 2.5, 3500)
x = sawtooth(np.pi * tn + np.pi/2, 0.5)

figure, ax = plt.subplots(1,figsize=(15, 8))
ax.xaxis.set_label_coords(1, .35)
ax.set_xlabel('t')
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.set_ylim(-1, 1)
ax.plot(tn, x)
ax.plot(tn[2000:3000], -x[2000:3000], ls='--')

ax.set_xticks([1, 2])
ax.set_xticklabels(["$T/2$", "$T$"])
ax.set_title("$x(t)$")

figure.show()
```

```
tn = np.linspace(-0.5, 1.5, 3500)
xr = np.sin(2 * np.pi * tn * 3)
x = np.sign(np.sin(2 * np.pi * tn))

figure, ax = plt.subplots(1,figsize=(15, 8))
ax.xaxis.set_label_coords(1, .35)
ax.set_xlabel('t')
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.set_ylim(-1, 1)
ax.plot(tn, x)
ax.plot(tn, xr)
ax.plot(tn[2000:3000], -x[2000:3000], ls='--')
ax.fill_between(tn, 0, xr, where=x*xr>=0, color="b")
ax.fill_between(tn, 0, xr, where=x*xr<0, color="r")
ax.set_xticks([0.5, 1])
ax.set_xticklabels(["$T/2$", "$T$"])
ax.set_title("$x(t)$")

figure.show()
```

## Simetrí a de cuarto de onda par.

Una señal $x(t)$, cuenta con simetría de cuarto de onda par si:



1.   La señal cuenta con simetría de median onda $x(t) = -x\left(t \pm \frac{T}{2}\right)$.

2.   La señal es par $x(t) = x(-t)$.






```
from scipy.signal import square

tn = np.linspace(-1.5, 2, 3500)
x = square(np.pi * tn + np.pi/2)

figure, ax = plt.subplots(1,figsize=(15, 8))
ax.xaxis.set_label_coords(1, .35)
ax.set_xlabel('t')
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.plot(tn, x, lw=3, c='k')
ax.set_xticks([-1, 1, 2])
ax.set_xticklabels(["$-T/2$", "$T/2$", "$T$"])
ax.set_title("$x(t)$")
for i in range(8):
    ax.axvline(-1.5 + 0.5*i, ls=":", c = "k")
ax.fill_between(tn, x, where=np.sin(np.pi*tn)>0, fc="cyan")
ax.fill_between(tn, x, where=np.sin(np.pi*tn)<0, fc="lime")
figure.show()
```

## Simetría de cuarto de onda impar.

Una señal $x(t)$ cuenta con simetría de cuarto de onda impart cuando:



1.   La señal cuenta con simetría de median onda $x(t) = -x\left(t \pm \frac{T}{2}\right)$.

2.   La señal es impar $x(t) = -x(-t)$.

```
tn = np.linspace(-2, 2.5, 4500)
x = sawtooth(np.pi * tn + np.pi/2, 0.5)

figure, ax = plt.subplots(1,figsize=(15, 8))
ax.xaxis.set_label_coords(1, .35)
ax.set_xlabel('t')
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.plot(tn, x, lw=3, c='k')
ax.set_xticks([-1, 1, 2])
ax.set_xticklabels(["$-T/2$", "$T/2$", "$T$"])
ax.set_title("$x(t)$")
ax.fill_between(tn, x, where=np.sin(np.pi*tn)>0, fc="cyan")
ax.fill_between(tn, x, where=np.sin(np.pi*tn)<0, fc="lime")
figure.show()
```

El análisis de los armónicos (serie trigonométrica de Fourier) teniendo en cuenta la simetría de media onda para $x(t)$, es decir: $x(t) = -x\left(t \pm \frac{T}{2}\right)$.

$$
a_0 = 0
$$

$$
a_n = \frac{2}{T} \int_{-T/2}^{T/2} x(t)\cdot cos(\omega_0\cdot n \cdot t)~dt\\
a_n = \frac{2}{T} \left[\int_{-T/2}^{0} x(t)\cdot cos(\omega_0\cdot n \cdot t)~dt+\int_{0}^{T/2} x(t)\cdot cos(\omega_0\cdot n \cdot t)~dt \right]
$$

Realizando el cambio de variable $t = \tau - \frac{T}{2}$, $dt = d\tau$.

$$
a_n = \frac{2}{T} \left[\int_{0}^{T/2} x\left(\tau - \frac{T}{2}\right)\cdot cos\left( \omega_0\cdot n \cdot \left(\tau - \frac{T}{2}\right) \right)~d\tau+\int_{0}^{T/2} x(t)\cdot cos(\omega_0\cdot n \cdot t)~dt \right]
$$

$$
cos(\omega_0\cdot n \cdot(\tau -\frac{T}{2})) = cos(\omega_0\cdot n \cdot \tau) \cdot cos(\omega_0\cdot n \cdot \frac{T}{2}) +  sin(\omega_0\cdot n \cdot \tau) \cdot sin(\omega_0\cdot n \cdot \frac{T}{2})\\
=  cos(\omega_0\cdot n \cdot \tau) \cdot cos(\frac{2\pi}{T}\cdot n \cdot \frac{T}{2}) +  sin(\omega_0\cdot n \cdot \tau) \cdot sin(\frac{2\pi}{T}\cdot n \cdot \frac{T}{2})\\
= cos(\omega_0\cdot n \cdot \tau) \cdot cos(\pi \cdot n) +  sin(\omega_0\cdot n \cdot \tau) \cdot sin(\pi \cdot n)\\
= cos(\omega_0\cdot n \cdot \tau) \cdot cos(\pi \cdot n)
$$

$$
a_n = \frac{2}{T} \left[\int_{0}^{T/2} -x(\tau) \cdot cos(\omega_0\cdot n \cdot \tau) \cdot cos(\pi \cdot n)~d\tau+\int_{0}^{T/2} x(t)\cdot cos(\omega_0\cdot n \cdot t)~dt \right]
$$

$$
a_n = \frac{2}{T} \int_{0}^{T/2} x(t)\cdot cos(\omega_0\cdot n \cdot t)~dt \cdot (1 - cos(\pi \cdot n))
$$

Por lo tanto:

$$
a_n =  \left\{
\begin{array}{ll}
      0 & n~par \\
      \frac{4}{T} \int_{0}^{T/2} x(t)\cdot cos(\omega_0\cdot n \cdot t)~dt & n~impar \\
\end{array}
\right. $$

$$
a_{2n-1} = \frac{4}{T} \int_{0}^{T/2} x(t)\cdot cos(\omega_0\cdot (2n-1) \cdot t)~dt\\
a_{2n} = 0
$$


Aplica lo mismo para $b_n$.

$$
b_n =  \left\{
\begin{array}{ll}
      0 & n~par \\
      \frac{4}{T} \int_{0}^{T/2} x(t)\cdot sen(\omega_0\cdot n \cdot t)~dt & n~impar \\
\end{array}
\right. $$

$$
b_{2n-1} = \frac{4}{T} \int_{0}^{T/2} x(t)\cdot sen(\omega_0\cdot (2n-1) \cdot t)~dt\\
b_{2n} = 0
$$

## Análisis pra armónicos para señales con simetría de cuarto de onda par.

Dada una señal $x(t)$ con simetria de media onda y simetria par.

$a_0 = 0$ por contar con simetría de media onda.

$b_n = 0$ por ser una señal par.

$a_{2n} = 0$ por ser una señal con simetría de media onda.

$a_{2n-1} = \frac{4}{T} \int_{0}^{T/2} x(t)\cdot cos(\omega_0\cdot (2n-1) \cdot t)~dt$

$$
a_{2n-1} = \frac{4}{T} \int_{0}^{T/4} x(t)\cdot cos(\omega_0\cdot (2n-1) \cdot t)~dt + \frac{4}{T} \int_{T/4}^{T/2} x(t)\cdot cos(\omega_0\cdot (2n-1) \cdot t)~dt
$$

Realizando el cambio de variable $t = \tau + \frac{T}{2}$ y $dt = d\tau$ en el segundo termimo.

$$
a_{2n-1} = \frac{4}{T} \int_{0}^{T/4} x(t)\cdot cos(\omega_0\cdot (2n-1) \cdot t)~dt + \frac{4}{T} \int_{-T/4}^{0} x(\tau + T/2)\cdot cos(\omega_0\cdot (2n-1) \cdot (\tau + T/2))~dt
$$

Ahora:

$$
cos(\omega_0\cdot (2n-1) \cdot (\tau +\frac{T}{2})) = cos(\omega_0\cdot (2n-1) \cdot \tau) \cdot cos(\omega_0\cdot (2n-1) \cdot \frac{T}{2}) +
sen(\omega_0\cdot (2n-1) \cdot \tau) \cdot sen(\omega_0\cdot (2n-1) \cdot \frac{T}{2})
$$

* $$ cos\left(\frac{2\pi}{T} \cdot (2n-1) \cdot \frac{T}{2}\right) = -1
$$

* $$ sen\left(\frac{2\pi}{T} \cdot (2n-1) \cdot \frac{T}{2}\right) = 0
$$

$$
cos(\omega_0\cdot (2n-1) \cdot (\tau +\frac{T}{2})) = -cos(\omega_0\cdot (2n-1) \cdot \tau)
$$

$$
a_{2n-1} = \frac{4}{T} \int_{0}^{T/4} x(t)\cdot cos(\omega_0\cdot (2n-1) \cdot t)~dt - \frac{4}{T} \int_{-T/4}^{0} x(\tau + T/2)\cdot cos(\omega_0\cdot (2n-1) \cdot \tau)~dt
$$

$$
a_{2n-1} = \frac{4}{T} \int_{0}^{T/4} x(t)\cdot cos(\omega_0\cdot (2n-1) \cdot t)~dt + \frac{4}{T} \int_{-T/4}^{0} x(\tau)\cdot cos(\omega_0\cdot (2n-1) \cdot \tau)~dt
$$

$$
a_{2n-1} = \frac{4}{T} \int_{-T/4}^{T/4} x(t)\cdot cos(\omega_0\cdot (2n-1) \cdot t)~dt
$$

Como $x(t)$ es par:

$$
a_{2n-1} = \frac{8}{T} \int_{0}^{T/4} x(t)\cdot cos(\omega_0\cdot (2n-1) \cdot t)~dt
$$

En resumen, una señal con simetria de cuarto de onda par cueta:

$a_0 = 0$ por contar con simetría de media onda.

$b_n = 0$ por ser una señal par.

$a_{2n} = 0$ por ser una señal con simetría de media onda.

$a_{2n-1} = \frac{8}{T} \int_{0}^{T/4} x(t)\cdot cos(\omega_0\cdot (2n-1) \cdot t)~dt$ por contar con simetría de cuarto de onda par.

## Análisis de armónicos se señales con simetría de cuarto de onda impar.

$x(t)$ es una señal con simetría de cuarto de onda impar. Por lo tanto al ser descrita a partir de su representación en series de Fourier se tiene:

$a_0 = 0$ por contar con simetría de media onda.

$a_n = 0$ por ser una señal impar.

$b_{2n} = 0$ por ser una señal con simetría de media onda.

$b_{2n-1} = \frac{8}{T} \int_{0}^{T/4} x(t)\cdot sen(\omega_0\cdot (2n-1) \cdot t)~dt$ por contar con simetría de cuarto de onda impar.

## Análisis de señales con simetría escondida.

Son señales con simetría impar que cuentan con un componente de frecuencia cero, que hace que pierda la citada simetría.

Dado $x(t)$, esta señal puede ser descompuesta en la suma de una señal DC más una señal con simetría impar.

$$
x(t) = C_0 + g(t)
$$

Para la señal $x(t)$ determine su simetría y calcule su representación utilizando el espacio trigonométrico de Fourier.

```
tn = np.linspace(-1, 3, 1000)
x = square(np.pi * tn)

figure = plt.figure(figsize=(15, 8))
ax = figure.add_subplot()
ax.plot(tn, x)
ax.xaxis.set_label_coords(1, .4)
ax.set_title("$x(t)$")
ax.set_xlabel('t')
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

figure.show()
```

Análisis de simetría.




*   La señal cuenta con simetría de onda impar porque $x(t) = -x(-t)$.
*   La señal cuenta con simetría de media onda porque $x(t) = -x\left(t \pm \frac{T}{2}
\right)$.
*   Finalmente, la señal cuenta con simetria de cuarto de onda impar.

$$
a_0 = 0
$$

$$
a_n = 0
$$

$$
b_{2n} = 0
$$

$$
b_{2n - 1} = \frac{8}{T} \int_{0}^{T/4} x(t)\cdot sen(\omega_0\cdot (2n - 1) \cdot t)~dt\\
= \frac{8}{T} \int_{0}^{T/4}sen(\omega_0\cdot (2n - 1) \cdot t)~dt\\
= \frac{4}{(2n-1)\cdot \pi}
$$

$$
x(t) = \sum_{n = 1}^{\infty} \frac{4}{(2n-1)\cdot \pi} \cdot sen(\omega_0\cdot (2n - 1) \cdot t)
$$


```
def Trigonometrica_Fourier(N,x,tn):
    import numpy as np
    from scipy.integrate import simps

    an =[]
    bn =[]
    xf = 0
    T0 = tn[-1] - tn[1]

    a0 = 2 * simps(x, tn) / T0
    xf = a0/2

    for i in range(1, N+1):
        integral_coseno = simps(x * np.cos(2 * np.pi * i * tn / T0), tn)
        an.append(integral_coseno*2/T0)

        integral_seno = simps(x * np.sin(2 * np.pi * i * tn / T0), tn)
        bn.append(integral_seno*2/T0)

        xf += an[i-1] * np.cos(2 * np.pi * i * tn / T0) + bn[i-1] * np.sin(2 * np.pi * i * tn / T0)

    return a0, an, bn, xf
```

```
tn = np.linspace(0, 2, 10000)
x = square(np.pi * tn)
N = 30
a0, an, bn, xf = Trigonometrica_Fourier(N, x, tn)
n = np.arange(1, N+1)

fig = plt.figure(figsize=(18, 9))

ax1 = fig.add_axes([0, 0, 0.45, 0.95])
ax2 = fig.add_axes([0.5, 0.5, 0.45, 0.45])
ax3 = fig.add_axes([0.5, 0, 0.45, 0.45])

ax1.plot(tn, x, label="Señal original")
ax1.plot(tn, xf, label="Señal reconstruida")
ax1.legend(loc="best")
ax1.grid()
ax1.margins(x=0)
ax1.set_xlabel("Tiempo")
ax1.set_title("Señal reconstruida utilizando {} armónicos".format(N))

ax2.stem(n ,np.around(an, 2), use_line_collection=True)
ax2.grid()
ax2.set_ylabel("Magnitud")
ax2.set_title("Coeficientes $a_n$")
ax2.set_xticks(n)
ax2.set_xticklabels([])

ax3.stem(n, bn, use_line_collection=True)
ax3.grid()
ax3.set_xlabel("Multiplos de la frecuencia fundamental")
ax3.set_ylabel("Fase (Grados)")
ax3.set_title("Coeficientes $b_n$")
ax3.set_xticks(n)

fig.show()
```

Análisis de simetría.

La señal $x(t)$ es una señal par, dado que $x(t) = x(-t)$. Tiene simetria de media onda dado que $x(t) = -x\left(t \pm \frac{T}{2}
\right)$. Por lo tanto:

$$
a_0 = 0
$$

$$
b_n = 0
$$

$$
2_{2n} = 0
$$

$$
a_{2n - 1} = \frac{8}{T} \int_{0}^{T/4} x(t)\cdot cos(\omega_0\cdot (2n - 1) \cdot t)~dt\\
= \frac{8}{T} \int_{0}^{T/4} cos(\omega_0\cdot (2n - 1) \cdot t)~dt\\
= \frac{4}{(2n-1)\cdot \pi} \cdot sen \left((2n-1) \cdot \frac{\pi}{2} \right)\\
a_{2n - 1} =  \left\{
\begin{array}{ll}
      \frac{4}{(2n-1)\cdot \pi} & para ~ (2n-1) = 1, 5, 9, ...\\
      -\frac{4}{(2n-1)\cdot \pi} & para ~ (2n-1) = 3, 7, 11, ...\\
\end{array}
\right.
$$

Por lo tanto $x(t)$ quedará representada así:

$$
x(t) = \sum_{n = 1}^{\infty} \frac{4}{(2n-1)\cdot \pi} \cdot (-1)^{n-1} \cdot cos(\omega_0\cdot (2n - 1) \cdot t)
$$

Teniendo en cuenta la simetría calcular la representación de $x(t)$ en los ejercicios pasados, utilizando la forma compleja de Fourier.

```
def Compleja_Fourier(N, x, tn):

    import numpy as np
    from scipy.integrate import simps

    C = []
    xf = 0
    T0 = tn[-1] - tn[0]

    for i in range(-N, N+1):
        sg = np.exp(1j*2*np.pi*i*tn/T0)
        C.append(simps(x*np.conjugate(sg), tn)/T0)
        xf += C[i+N] * sg

    return C, xf
```

```
tn = np.linspace(0, 2, 10000)
x = square(np.pi * tn)
N = 10
C, xf = Compleja_Fourier(N, x, tn)
n = np.arange(-(len(C)-1)/2, (len(C)+1)/2)

fig = plt.figure(figsize=(18, 9))

ax1 = fig.add_axes([0, 0, 0.45, 0.95])
ax2 = fig.add_axes([0.5, 0.5, 0.45, 0.45])
ax3 = fig.add_axes([0.5, 0, 0.45, 0.45])

ax1.plot(tn, x, label="Señal original")
ax1.plot(tn, np.real(xf), label="Señal reconstruida")
ax1.legend(loc="best")
ax1.grid()
ax1.margins(x=0)
ax1.set_xlabel("Tiempo")
ax1.set_title("Señal reconstruida utilizando {} armónicos".format(N))

ax2.stem(n ,np.abs(C), use_line_collection=True)
ax2.grid()
ax2.set_ylabel("Magnitud")
ax2.set_title("Coeficientes $C_n$")
ax2.set_xticks(n)
ax2.set_xticklabels([])

ax3.stem(n, np.angle(C, True), use_line_collection=True)
ax3.grid()
ax3.set_xlabel("Multiplos de la frecuencia fundamental")
ax3.set_ylabel("Fase (Grados)")

ax3.set_xticks(n)

fig.show()
```
